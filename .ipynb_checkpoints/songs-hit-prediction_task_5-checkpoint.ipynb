```python
#Importing Necessary Libraries
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

import warnings
warnings.filterwarnings(action='ignore')
```

```python
#Importing Datasets at once and string them in a list dfs(Dataframes)
dfs = [pd.read_csv(f'input/the-spotify-hit-predictor-dataset/dataset-of-{decade}0s.csv') for decade in ['6', '7', '8', '9', '0', '1']]
```

```python
#Gething first dataset 
dfs[0]
```

```python
#Gething all datasets
dfs
```

```python
#Adding a new column Decade for each of our datasets with coressponding order.

for i, decade in enumerate([1960, 1970, 1980, 1990, 2000, 2010]):
    dfs[i]['decade'] = pd.Series(decade, index=dfs[i].index)
```

```python
#As you can see at the end we have decade colunm
dfs[0]
```

```python
#now merging all the datasets and with axis 0 (means top of each other) and performing random shuffle on decade column also 
#resetting the index of rows and stroing in variable data.

data = pd.concat(dfs, axis=0).sample(frac=1.0, random_state=1).reset_index(drop=True)
```

```python
#now all the datasets are according to my thoughts
data
```

```python
#Checking for the missing value before the preprocessing
data.info()
```

```python
def preprocess_inputs(df):
    df = df.copy()
    
    # Drop high-cardinality categorical columns because its not going to help our model
    df = df.drop(['track', 'artist', 'uri'], axis=1)
    
    # Split df into X and y
    y = df['target']
    X = df.drop('target', axis=1)
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    # Scale X
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
    
    return X_train, X_test, y_train, y_test
```

```python
#Giving input which is our dataset to our function stroing results in thier return values 
X_train, X_test, y_train, y_test = preprocess_inputs(data)
```

Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define the hyperparameters to search over
param_grid = {
    'n_estimators': [10, 100, 500],
    'max_depth': [2, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}
```

```python
# Create a RandomForestClassifier model with default hyperparameters
rf = RandomForestClassifier()
```

```python
# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
# Fit the grid search to your data
grid_search.fit(X_train, y_train)
```

```python
# Print the best hyperparameters found
print('Best hyperparameters:', grid_search.best_params_)
```
